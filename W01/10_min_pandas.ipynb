```python
'''
By â€œgroup byâ€ we are referring to a process involving one or more of the following steps:

* Splitting the data into groups based on some criteria
* Applying a function to each group independently
* Combining the results into a data structure
'''

df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three',
                          'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})
```

```python
df
```

```python
#Grouping and then applying the sum() function to the resulting groups.
df.groupby('A').sum()
```

```python
#Grouping by multiple columns forms a hierarchical index, and again we can apply the sum function.
df.groupby(['A', 'B']).sum()
```

```python
#Reshaping
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                      'foo', 'foo', 'qux', 'qux'],
                      ['one', 'two', 'one', 'two',
                       'one', 'two', 'one', 'two']]))
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
```

```python
df2 = df[:4]
```

```python
df2
```

```python
#The stack() method â€œcompressesâ€ a level in the DataFrameâ€™s columns
stacked = df2.stack()
```

```python
stacked
```

```python
#With a â€œstackedâ€ DataFrame or Series (having a MultiIndex as the index), 
#the inverse operation of stack() is unstack(), 
#which by default unstacks the last level
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```
